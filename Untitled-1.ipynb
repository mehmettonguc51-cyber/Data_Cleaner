```python
# Dataset'teki TÜM tabloları listele ve her birini değişkene ata
from google.cloud import bigquery
import pandas as pd

PROJECT_ID = "datawarehouseanalytics-480618"
DATASET_ID = "datawarehouseset"

client = bigquery.Client(project=PROJECT_ID)

# 1. Dataset içindeki tüm tablo isimlerini al
dataset_ref = f"{PROJECT_ID}.{DATASET_ID}"
tables_list = client.list_tables(dataset_ref)

table_names = [table.table_id for table in tables_list]
print(f"Bulunan tablolar: {table_names}")

# 2. Her tabloyu bir dictionary'de sakla (tables["tablo_adi"] = DataFrame)
tables = {}
for name in table_names:
    full_table_id = f"{PROJECT_ID}.{DATASET_ID}.{name}"
    tables[name] = client.list_rows(full_table_id).to_dataframe()
    print(f"✓ {name}: {tables[name].shape[0]} satır, {tables[name].shape[1]} sütun")

# Kullanım örnekleri:
# tables["gold_dim_products"]   -> products tablosu
# tables["gold_dim_customers"] -> customers tablosu (varsa)
# veya: df = tables["tablo_adi"]
```

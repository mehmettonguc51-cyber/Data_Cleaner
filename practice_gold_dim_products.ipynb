## 1. Ortam Kurulumu

```python
# Gerekli kütüphaneler
from google.cloud import bigquery
import pandas as pd
import numpy as np
from pathlib import Path

# Görselleştirme (opsiyonel)
# import matplotlib.pyplot as plt
# import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

# BigQuery config
PROJECT_ID = "datawarehouseanalytics-480618"
DATASET_ID = "datawarehouseset"
client = bigquery.Client(project=PROJECT_ID)
```

## 2. Veri Yükleme - gold_dim_products → df

```python
# BigQuery'dan gold_dim_products tablosunu çek
TABLE_NAME = "gold_dim_products"
full_table_id = f"{PROJECT_ID}.{DATASET_ID}.{TABLE_NAME}"
query = f"SELECT * FROM `{full_table_id}`"
df = client.query(query).to_dataframe()

print(f"Satır: {len(df)}, Sütun: {len(df.columns)}")
df.head()
```
